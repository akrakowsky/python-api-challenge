```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os

# Import API key
from config import g_key
```

### Store Part I results into DataFrame
* Load the csv exported in Part I to a DataFrame

```python
#R Read the csv file
file_read = pd.read_csv("./Cities_Output.csv")

# Create a data frame
vacation_data = pd.DataFrame(file_read) 
vacation_data
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
# Configure gmaps
gmaps.configure(api_key= g_key)
```

```python
# Create a heat map for humidity per location

# Store Lat and Lng
locations = vacation_data[["Lat", "Lng"]]

# Replace Nan values and convert to float
humidity = vacation_data["Humidity"].astype(float)

# Plot Heatmap
fig = gmaps.figure()

# Create heat layer
heat_layer = gmaps.heatmap_layer(locations, weights=humidity, 
                                 dissipating=False, max_intensity=10,
                                 point_radius=1)


# Add layer
fig.add_layer(heat_layer)

# Display figure

fig
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
# Narrow down cities that fit criteria and drop any results with null values
vacation_df = vacation_data.loc[(vacation_data['Max Temp'] > 65) & (vacation_data['Wind Speed'] < 10) & 
                                (vacation_data['Cloudiness'] < 10),:]
len(vacation_df)
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
# Create the hotel data frame the finds hotels within 5000 meters of ideal locations
hotel_df = vacation_df[['City','Country','Lat','Lng']].reset_index(drop=True)
hotel_df.head()
```

```python
# Add Hotel Name as new column
hotel_df["Hotel Name"] = ""
hotel_df. head()
```

```python
# Set parameters and find hotels within 5000 meters
params = {
    "radius": 5000,
    "keyword": "lodging",
    "key": g_key,
}

# Search through list
for index, row in hotel_df.iterrows():
    # get lat, lng from df
    lat = row["Lat"]
    lng = row["Lng"]

    # Get locations
    params["location"] = f"{lat},{lng}"
    
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

    # Request
    print(f"Retrieving Results for City {index+1}: {row['City']}.")
    response = requests.get(base_url, params=params).json()
    # print(json.dumps(response, indent=4, sort_keys=True))
    
    # Find results
    results = response['results']
    
    try:
        print(f"Lodging option is {results[0]['name']}.")
        hotel_df.loc[index, 'Hotel Name'] = results[0]['name']
        
    except (KeyError, IndexError):
        print("Skipping.")
        
    print("------------")
```

```python
# Remove the hotels with missing criteria
hotel_df.dropna()
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
locations = hotel_df[["Lat", "Lng"]]

markers = gmaps.marker_layer(locations)
vacation_spots = gmaps.symbol_layer(locations, info_box_content = hotel_info, scale = 2)
```

```python
# Add marker layer ontop of heat map

fig.add_layer(markers)
fig.add_layer(vacation_spots)

# Display figure
fig
```
